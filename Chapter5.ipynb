```python
car = 'subaru'
print("Is car == 'subaru'? I predict True.")
print(car == 'subaru')  # True

print("Is car == 'audi'? I predict False.")
print(car == 'audi')    # False
```

```python
# Test 1
print("Is 5 > 3? I predict True.")
print(5 > 3)

# Test 2
print("Is 'apple' == 'Apple'? I predict False.")
print('apple' == 'Apple')

# Test 3
print("Is 10 != 9? I predict True.")
print(10 != 9)

# Test 4
print("Is 5 <= 3? I predict False.")
print(5 <= 3)

# Test 5
print("Is 'a' < 'b'? I predict True.")
print('a' < 'b')
```

```python
# Equality and Inequality with strings
name = "Sufyan"
print(name == "Sufyan")
print(name != "Ali")

# Using lower()
print(name.lower() == "sufyan")

# Numerical tests
age = 20
print(age == 20)
print(age != 25)
print(age > 18)
print(age < 30)
print(age >= 20)
print(age <= 20)

# Using and/or
print(age > 18 and age < 25)
print(age > 18 or age < 10)

# In and not in
items = ['apple', 'banana', 'orange']
print('banana' in items)
print('grape' not in items)
```

```python
usernames = ['admin', 'sufyan', 'hamza', 'ali', 'zain']

for user in usernames:
    if user == 'admin':
        print("Hello admin, would you like to see a status report?")
    else:
        print(f"Hello {user}, thank you for logging in again.")
```

```python
current_users = ['Ali', 'Hamza', 'Zain', 'Admin', 'Asad']
new_users = ['zain', 'Sufyan', 'Usman', 'admin', 'Hamid']

current_users_lower = [user.lower() for user in current_users]

for new_user in new_users:
    if new_user.lower() in current_users_lower:
        print(f"Sorry {new_user}, that username is taken.")
    else:
        print(f"{new_user} is available.")
```

```python
numbers = list(range(1, 10))

for number in numbers:
    if number == 1:
        print("1st")
    elif number == 2:
        print("2nd")
    elif number == 3:
        print("3rd")
    else:
        print(f"{number}th")
```
