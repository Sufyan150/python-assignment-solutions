```python
glossary = {
    'variable': 'A name used to store data.',
    'loop': 'A way to repeat actions in code.',
    'dictionary': 'A key-value pair data type.',
    'list': 'A collection of items in a particular order.',
    'string': 'A series of characters.'
}

for word, meaning in glossary.items():
    print(f"{word.title()}:\n  {meaning}\n")
```

```python
rivers = {
    'nile': 'egypt',
    'amazon': 'brazil',
    'indus': 'pakistan'
}

for river, country in rivers.items():
    print(f"The {river.title()} runs through {country.title()}.")

print("\nRivers mentioned:")
for river in rivers:
    print(river.title())

print("\nCountries mentioned:")
for country in rivers.values():
    print(country.title())
```

```python
favorite_languages = {
    'ali': 'python',
    'hamza': 'java',
    'zain': 'c++',
    'usman': 'python'
}

people = ['sufyan', 'ali', 'hamza', 'fatima']

for person in people:
    if person in favorite_languages:
        print(f"Thank you {person.title()} for taking the poll!")
    else:
        print(f"{person.title()}, please take the poll!")
```

```python
pet1 = {'animal': 'dog', 'owner': 'sufyan'}
pet2 = {'animal': 'cat', 'owner': 'ali'}
pet3 = {'animal': 'parrot', 'owner': 'hamza'}

pets = [pet1, pet2, pet3]

for pet in pets:
    print(f"{pet['owner'].title()} has a {pet['animal']}.")
```

```python
favorite_places = {
    'sufyan': ['murree', 'gilgit'],
    'ali': ['lahore'],
    'hamza': ['karachi', 'quetta', 'peshawar']
}

for person, places in favorite_places.items():
    print(f"{person.title()}'s favorite places:")
    for place in places:
        print(f" - {place.title()}")
```

```python
cities = {
    'islamabad': {
        'country': 'pakistan',
        'population': '1 million',
        'fact': 'Capital city of Pakistan.'
    },
    'karachi': {
        'country': 'pakistan',
        'population': '15 million',
        'fact': 'Largest city in Pakistan.'
    },
    'lahore': {
        'country': 'pakistan',
        'population': '11 million',
        'fact': 'Famous for historical sites.'
    }
}

for city, info in cities.items():
    print(f"\nCity: {city.title()}")
    for key, value in info.items():
        print(f"  {key.title()}: {value}")
```
