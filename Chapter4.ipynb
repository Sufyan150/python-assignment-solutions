```python
pizzas = ["pepperoni", "cheese", "bbq chicken"]
for pizza in pizzas:
    print(f"I like {pizza} pizza.")
print("I really love pizza!")
```

```python
animals = ["cat", "dog", "rabbit"]
for animal in animals:
    print(f"A {animal} would make a great pet.")
print("Any of these animals would make a great pet!")
```

```python
cubes = []
for number in range(1, 11):
    cubes.append(number ** 3)

for cube in cubes:
    print(cube)
```

```python
players = ["sufyan", "ali", "hassan", "usman", "hamza"]
print("The first three items in the list are:", players[:3])
print("Three items from the middle are:", players[1:4])
print("The last three items in the list are:", players[-3:])
```

```python
my_pizzas = ["pepperoni", "cheese", "veggie"]
friend_pizzas = my_pizzas[:]

my_pizzas.append("bbq chicken")
friend_pizzas.append("tikka")

print("My favorite pizzas are:")
for pizza in my_pizzas:
    print(pizza)

print("\nMy friend's favorite pizzas are:")
for pizza in friend_pizzas:
    print(pizza)
```
