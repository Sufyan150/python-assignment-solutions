```python
guests = ["Hassan", "Nashit", "Asad", "Tariq", "Sufyan", "Usman"]
print("I can invite only two people.")

while len(guests) > 2:
    removed = guests.pop()
    print(f"Sorry {removed}, I can't invite you.")

for guest in guests:
    print(f"{guest}, you're still invited.")

# guests.clear()
```

```python
places = ["Tokyo", "London", "New York", "Mecca", "Dubai"]
print("Original:", places)
print("Sorted:", sorted(places))
print("Original Again:", places)
print("Reverse Sorted:", sorted(places, reverse=True))
print("Original Again:", places)

places.reverse()
print("Reversed:", places)
places.reverse()
print("Original Again:", places)

places.sort()
print("Sorted Permanently:", places)
places.sort(reverse=True)
print("Reverse Sorted Permanently:", places)
```

```python
languages = ["Python", "Java", "C++", "Ruby", "JavaScript"]
print(languages)
print(languages[0])
print(languages[-1])
languages.append("Go")
languages.insert(1, "Swift")
print(languages)
del languages[2]
popped = languages.pop()
print(f"Popped language: {popped}")
languages.remove("Ruby")
languages.sort()
print("Sorted:", languages)
languages.sort(reverse=True)
print("Reverse Sorted:", languages)
languages.reverse()
print("Reversed:", languages)
print("Length of list:", len(languages))
```
